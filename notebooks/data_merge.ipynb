```python
import pandas as pd
```

```python
for y in range(2001,2022):
    print("\n",y)
    df1 = pd.read_csv(f"./data-samples/full/full{y}01.dat")#,dtype="str",encoding="utf-8",encoding_errors='replace',on_bad_lines='skip',lineterminator='\n')
    df1.to_parquet(f"./data-samples/movements/full{y}01.parquet")
    df = df1[df1.PRODUCT_NC == "TOTAL"]
    del df1
    for m in range(2,13):
        print(str(m)+" ",end="")
        df2 = pd.read_csv(f"./data-samples/full/full{y*100+m}.dat")#,dtype="str",encoding="utf-8",encoding_errors='replace',on_bad_lines='skip',lineterminator='\n')
        df2.to_parquet(f"./data-samples/movements/full{y*100+m}.parquet")
        df2 = df2[df2.PRODUCT_NC == "TOTAL"]
        df = pd.concat([df,df2])
        del df2
    df.reset_index(drop=True).to_parquet(f"./data-samples/full/Month/full{y}_total.parquet")
    del df
```
