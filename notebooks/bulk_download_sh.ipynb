```python
for f in tqdm(os.listdir("./tr")):
    if "7z" in f:
        if "52" in f:
            try:
                os.system(f'7z x -o"./tr/" "./tr/{f}"')
                os.system(f'rm -v "./tr/{f}"')
                pd.read_csv(f'./tr/{f.replace("7z","dat")}',encoding="latin",low_memory=False,na_filter=False).to_parquet(f"./tr/Years/{f.replace('.7z','.parquet')}")
                os.system(f'rm -v "./tr/{f.replace("7z","dat")}"') 
            except:
                print(f)
        else:
            try:
                os.system(f'7z x -o"./tr/" "./tr/{f}"')
                os.system(f'rm -v "./tr/{f}"')
                pd.read_csv(f'./tr/{f.replace("7z","dat")}',encoding="latin",low_memory=False,na_filter=False).to_parquet(f"./tr/Months/{f.replace('.7z','.parquet')}")
                os.system(f'rm -v "./tr/{f.replace("7z","dat")}"') 
            except:
                print(f)
```
